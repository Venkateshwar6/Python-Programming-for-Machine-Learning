```python
import pandas as pd
car_selling_data={'Brand':['Maruti','Maruti','Maruti','maruti','hyundai','hyundai','toyoto','mahentra','mahentra','forch','toyoto','forch'],'year':[2010,2011,2009,2010,2009,2009,2008,2010,2011,2010,2008,2010],'sold':[6,7,8,6,5,6,7,8,6,4,6,6]}
df=pd.DataFrame(car_selling_data,columns=['Brand','year','sold'])
print(df)
grouped=df.groupby('year')
print("Group by year\n",grouped.get_group(2010))
```
